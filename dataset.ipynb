# Dataset

### datasets overview
for this project we have chosen a dataset called German Credit Risk.

the original dataset is available here: `https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data`. It contains 1000 entries with 20 categorial/symbolic attributes prepared by Prof. Hofmann.

note - on the day of writing that report (10.04.2026) this dataset is unavailable - the site gives error 502

However that dataset is difficult to understand due to its complicated system of categories and symbols. So the user UCI Machine Learning created and published on Kaggle more readible version of it: `https://www.kaggle.com/datasets/uciml/german-credit/data`


### dataset attributes
We decided to use the Kaggle version of the dataset. It consisted of the following 10 attributes (that are self explaining):
- Age (numeric)
- Sex (text: male, female)
- Job (numeric: 0 - unskilled and non-resident, 1 - unskilled and resident, 2 - skilled, 3 - highly skilled)
- Housing (text: own, rent, or free)
- Saving accounts (text - little, moderate, quite rich, rich)
- Checking account (numeric, in DM - Deutsch Mark)
- Credit amount (numeric, in DM)
- Duration (numeric, in month)
- Purpose (text: car, furniture/equipment, radio/TV, domestic appliances, repairs, education, business, vacation/others)



### dataset problems
The dataset had some issues we had to fix.

1. on Kaggle there was no decision attribute (final class) so we merged it with original datasets decision attribute (code available in `datasset_refining` folder)

2. there were some missing values in the Kaggle dataset, which were not present in the original dataset. After some time and research we deduced what was the reason for it and we filled dataset with appropriate values

### monotonicity of attributes for NN-UTA-DIS

we had to verify the monotonicity of the attributes for the NN-UTA-DIS method. It was not obvious for some of the attributes if the monotonicity exist and for others we had to change the labeling of some attributes into more meaningful pattern. We also checked the corellation of some attributes with the final class to enable for more reasonable ordering of values

The final attributes we used for that method (before normalisation) were:
- `Age` (numeric) — kept as-is; older age correlates with higher creditworthiness (monotonically increasing with credit risk score)
- `Saving accounts` (re-encoded: little=0, moderate=1, quite rich=2, rich=3) — more savings indicates lower risk, so the ordering is monotonically increasing
- `Checking account` (re-encoded: little=1, moderate=2, rich=3; missing values filled with 3) — similarly, higher balance indicates lower risk; missing values were filled with 3 as their absence typically indicates companies with strong financial standing
- `Credit amount` (negated: stored as negative value) — higher credit amount correlates with higher risk, so we negated it to maintain monotonic increasing direction toward low risk
- `Duration` (negated: stored as negative value) — longer loan duration correlates with higher risk, so it was negated for the same reason
- `Purpose` (re-encoded: car/radio/TV/vacation=0, furniture/repairs/domestic appliances=1, education/business=2) —  purposes were grouped into irresponsible, necessary, ambitious groups  (by their typical risk level, from lowest to highest expected risk)

Attributes removed:
- `Sex` — dropped to avoid discrimination-related bias
- `Housing` and `Job` — dropped due to unclear or non-monotonic relationship with credit risk
